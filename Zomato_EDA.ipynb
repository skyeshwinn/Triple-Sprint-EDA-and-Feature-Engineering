```python
df_country = pd.read_excel('C:/Users/skyes/OneDrive/Documents/Learning/EDA and Feature Engineering/1. Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
#Merging the two dataframes on the common column 'Country Code'
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
country_names = final_df.Country.value_counts().index
```

```python
## Pie Chart - Top 3 countries with the most number of records in the dataset
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

Observations: 

Aggregate rating is between 4.5-4.9 --> Excellent

Aggregate rating is between 4.0-4.4 --> Very Good

Aggregate rating is between 3.5-3.9 --> Good

Aggregate rating is between 2.5-3.4 --> Average

Aggregate rating is between 0.0-2.4 --> Poor

```python
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color', palette=['grey', 'red', 'orange', 'yellow', 'lime', 'green'])
plt.show()
```

Observations:

'Not rated' count is very high.

Maximum no. of ratings are between 2.5-3.4.

```python
#Count plot
sns.countplot(x='Rating color', data=ratings, palette=['grey', 'red', 'orange', 'yellow', 'lime', 'green'])
plt.show()
```

```python
#Count plot to show the number of unique ratings(frequency) in the dataset
sns.countplot(x='Rating color', data=ratings, hue = 'Rating color', palette=['grey', 'red', 'orange', 'yellow', 'lime', 'green'], legend=False)
plt.show()
```
